# 2. import dataset

```python
df = pd.read_csv('../inputs/2_bins_financial_grade.csv')
df = df[['Fund profile: Fund name', 'Fund profile: Category group', 'Fund profile: Fund net assets', 'Grade', 'Financial performance']]
df = df.sort_values('Fund profile: Fund net assets', ascending=False)
df.shape
```

```python
df1 = df.groupby(['Financial performance','Grade']).agg({'Fund profile: Fund name':'count'}).reset_index()
df1
```

# 3. Top-Mid

```python
df1 = df[df['Financial performance']=='a. top']
df1 = df1[df1['Grade']=='b. mid']

df2 = df[df['Financial performance']=='b. mid']
df2 = df2[df2['Grade']=='a. top']

df3 = pd.concat([df1,df2])
df3 = df3.sort_values('Fund profile: Fund net assets', ascending=False)
df3
```
